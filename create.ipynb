```python
import pandas as pd
import numpy as np
import csv
import time
from datetime import datetime
#Thư viện để dừng schedule
import sys
import os
import signal

def Create(mini, maxi, op, cl):
    tim_nen = abs(float(maxi) - float(mini))
    low_per = abs(((float(op) - float(mini))/tim_nen)*100)
    high_per = abs(((float(cl) - float(mini))/tim_nen)*100)
    print(low_per, "----",tim_nen, high_per)

    if high_per > low_per:
        if low_per <= 5 and high_per >= 95:
            print(low_per, "----", high_per)
            return 2
        elif low_per >= 35 and low_per <= 40 and high_per >= 60 and high_per <= 65:
            print(low_per, "----", high_per)
            return 3
        elif low_per <= 5 and high_per >= 25 and high_per >= 45:
            print(low_per, "----", high_per)
            return 4
        elif low_per >= 55 and low_per <= 75 and high_per >= 95: 
            print(low_per, "----", high_per)
            return 5
        else:
            print(low_per, "----", high_per)
            return 1

    elif high_per < low_per:
        if low_per >= 95 and high_per <= 5:
            print(low_per, "----", high_per)
            return 7
        elif low_per >= 60 and low_per <= 65 and high_per >= 35 and high_per <= 40:
            print(low_per, "----", high_per)
            return 8
        elif low_per >= 25 and low_per <= 45 and high_per <= 5:
            print(low_per, "----", high_per)
            return 9
        elif low_per >= 95 and high_per <= 55 and high_per >= 75:
            print(low_per, "----", high_per)
            return 10
        else:
            print(low_per, "----", high_per)
            return 6
    elif abs(high_per - low_per) <= 2:
        return 11
    else:
        print("N")


```

```python
Ma_chung_khoan=open("Ma_chung_khoan.txt","r")
Ma_CK=Ma_chung_khoan.readlines()
Ma_chung_khoan.close() 
#loại bỏ dấu  ","
Ma_CK=Ma_CK[0].split(',')
for mck in Ma_CK:
    #Đường link file cào, sys.path[0] là đường dẫn thư mục của code         
    save_path = sys.path[0].replace('\\','/')+'/NEN/'+mck
    #Kiểm tra xem đã có folder đó chưa, nếu chưa thì tạo mới
    if not os.path.exists(save_path):
        os.makedirs(save_path)
def xu_ly_file(mck):
    save_path = sys.path[0].replace('\\','/')+'/DATA_HOSE/'+mck
    st = save_path+"/"+mck+"_"+datetime.today().strftime('%d_%m_%Y')+".csv" 
    #st = save_path+"/"+"AAA_10_11_2021"+".csv"
    file = open(st,"r")
    data = pd.read_csv(st,delimiter=',')
    x = np.array(data)
    row_count = sum(1 for row in file)
    for i in range(0,row_count):
        if(x[i][11] != "N"):
            op = x[i][11]
            break
    mini = x[-1][23]
    maxi = x[-1][22]
    cl = x[-1][11]
    print("dong cua",cl)
    #Đường link file cào, sys.path[0] là đường dẫn thư mục của code 
    save_path = sys.path[0].replace('\\','/')+'/NEN/'+mck
    #Tên file.csv: Mã chứng khoán + ngày tháng năm
    completeName = os.path.join(save_path,mck+".csv")
    #sửa mode w+ thành mode a để viết được vào cuối file
    a=[]
    a.append(datetime.today().strftime('%d_%m_%Y'))
    a.append(str(Create(mini,maxi,op,cl)))
    file = open(completeName,"a",)
    with file:
        write = csv.writer(file)
        write.writerows(np.array(a).reshape(-1,2))
def ghi():
    for mck in Ma_CK:
        xu_ly_file(mck)
ghi()
```
